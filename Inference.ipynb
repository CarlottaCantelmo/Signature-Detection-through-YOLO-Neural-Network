```python
text_path = 'Yolov5/runs/detect/exp3/labels/*.txt'
path = 'Yolov5/runs/detect/exp3/labels'
text_names = []

for text in glob.glob(text_path, recursive=True):
    textname = os.path.basename(text)
    text_names.append(textname)
    #print(text_names)
    single_text = os.path.join(path, textname)
```

CREATE DATAFRAME FROM MODEL RESULTS: PREDICTED

```python
new_list = []
for file in Path(path).glob('*'):
    textname = os.path.basename(file)
    with open(file, "r") as prova:
        for line in prova:
            line = line + textname
            stripped_line = line.strip()
            line_list = stripped_line.split()
            new_list.append(line_list)
        prova.close()
    print(new_list)
```

```python
# Conversion of str to int #
stage1 = []
for i in range(0, len(new_list)):
    if i == int:
        test_list = list(map(float, new_list[i]))
    else:
        test_list = list(map(str, new_list[i]))
    stage1.append(test_list)
```

```python
new_dataframe = pd.DataFrame(stage1)
new_dataframe.columns = ["Classes","x_min","y_min","width","height","new_filename"]
new_dataframe
```

CREATE DATAFRAME FROM TEST IMAGES

```python
path_valid_l = 'New_format/labels/valid'
```

```python
labels_list = []
for file in Path(path_valid_l).glob('*'):
    labels = os.path.basename(file)
    with open(file, "r") as prova:
        for line in prova:
            line = line + labels
            stripped_line = line.strip()
            line_list = stripped_line.split()
            labels_list.append(line_list)
        prova.close()
    print(labels_list)
```

```python
prepare = []
for i in range(0, len(labels_list)):
    if i == int:
        test_list = list(map(float, labels_list[i]))
    else:
        test_list = list(map(str, labels_list[i]))
    prepare.append(test_list)

prepare
```

```python
df = pd.DataFrame(prepare)
df.columns = ["Classes","x_MIN","y_MIN","wid","heig","new_filename"]
df
```

```python
df['key'] = df['new_filename']+'-'+df['Classes'].astype(str)
```

```python
new_dataframe['key'] = new_dataframe['new_filename']+'-'+new_dataframe['Classes'].astype(str)
```

FINAL DATASET

```python
data = pd.merge(df, new_dataframe, on = ['key'], how = 'inner')
data
```
